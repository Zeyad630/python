## Comparison Operators
Comparison operators are used to compare two values:

```python
## ==	Equal

print(5 == 10) # False
print(5 == 5) # True


## !=	Not equal
print()
print(5 != 10) # True
print(5 != 5) # False

## >	Greater than

## <	Less than

## >=	Greater than or equal to
print()
print(12 >= 10)
## <=	Less than or equal to
```

## Logical Operators
Logical operators are used to combine conditional statements:

```python
## and (&)	Returns True if both statements are true
(5 == 5) and (10 == 10) # True
## or	(|) Returns True if one of the statements is true
(5 == 6) or (10 == 10) # True
## not	(~) Reverse the result, returns False if the result is true
not(5 > 11)
```

# Control Flow (if conditions)

In computer programming, the `if` statement is a conditional statement. It is used to execute a block of code only when a specific condition is met.

```python
# only students with GPA greater than 3.6 will get a gift.

gpa = 3.0

if gpa > 3.6:
  print("gift")
print('gazar')
```

```python
# student with 50 and above will pass, otherwise he will fail
grade = 40

if grade >= 50:
  print("pass") # true
else:
  print("fail") # false


## inline if / ternary operator
## true if(condition) else False
print("pass") if grade >= 50 else print("fail")
```

Suppose we need to assign different grades to students based on their scores.

> If a student scores above 90, assign grade A

> If a student scores above 75, assign grade B

> If a student scores above 65, assign grade C

```python
grade = 50

if grade > 90:
  print('A')
elif grade > 75:
  print('B')
elif grade > 65:
  print('C')
else:
  print('F')


```
